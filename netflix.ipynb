```python
data2 = pd.read_csv('shows.csv')
data2.head()
```

```python
cleaned_data2 = data2.dropna()
cleaned_data2.to_csv('netflix_shows_cleaned.csv', index=False)
```

Catégoriser la durée en groupes


```python
data2['runtime'] = data2['runtime'].str.extract('(\d+)').astype(float)
```

```python
bins = [0, 30, 60, float('inf')]
```

```python

labels = ['courte', 'moyenne', 'longue']  # Labels pour chaque groupe de durée

```

```python
data2['duration_category'] = pd.cut(data2['runtime'], bins=bins, labels=labels, right=False)
```

```python
data2.to_csv('netflix_shows_with_duration_category.csv', index=False)
```

Conversion de la qualité en format binaire :

```python
# Convertir la qualité en format binaire
data2['is_4k'] = data2['quality'].apply(lambda x: 1 if x == '4K' else 0)

# Enregistrer les données avec la conversion de qualité en format binaire dans un nouveau fichier CSV
data2.to_csv('netflix_shows_with_binary_quality.csv', index=False)
```
